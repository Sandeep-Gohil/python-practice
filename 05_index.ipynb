```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
# creating the new df frame where we are filtering the data on job counrty column to just have canada in it
# to do that I filtered job counry columne == canada so my new dataframe will only shows the records of job posting in caanada

df_canada = df[df['job_country']== 'Canada'] 
df_canada
```

```python
# lets say you have created the new dataframe using original data using filters it will have the same index order as original, means it won't be 
# starting from 0 to ... so to do that we will use .reset_index() to create new index while keeping the original as it is bcz in feature
# you might need merge the both data frame. 

df_canada.reset_index(inplace=True)

df_canada
```

```python
df_canada.set_index('job_posted_date', inplace=True) #.set_index will set the columns as dataframe index  
```
