```python
# Importing Libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

```

```python
#only get data analyst job in us

df_DA_US = df[(df['job_country'] == 'United States') & (df['job_title_short'] == 'Data Analyst')].copy()

df_DA_US 
```

```python
from matplotlib import ticker

df_DA_US['salary_year_avg'].plot(kind='hist', bins=40, edgecolor='black')
plt.xlim(0,250000)
# lets format x axis we did it in previous exercise 

ax = plt.gca()  #this will the axis then you can do modifications. like below
ax.xaxis.set_major_formatter(plt.FuncFormatter( lambda x, _: f'${int(x/1000)}K'))
plt.xlabel('Yearly Salary')
plt.ylabel('Number of Jobs')
plt.title('Distribution of US Data Analyst Yearly Salaries')
plt.show()
```
