```python
# Importing Libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt  

# Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
#General trend of data jobs in demand. We're going to look at it by month.
#First step is we're only going to look at roles in the United States. So we're going to
#  create a copy of our DataFrame and only include roles that are in the United States.

df_US = df[df['job_country'] == 'United States'].copy()

df_US
```

```python
#Since we're getting the count by month we need to aggregate by month.
# We can do this by first extracting the month from our job_posted_date column.

df_US['job_posted_month'] = pd.to_datetime(df_US['job_posted_date']).dt.strftime('%B')

df_US['job_posted_month']
```

```python
# let's pivot by month and job_title_short and get the count.

df_US_pivot =  df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')
```

```python
df_US_pivot
```

```python
# sort job_posted_month_name in df_US_pivot by job_posted_month value

df_US_pivot = df_US_pivot.reset_index()
df_US_pivot['job_posted_month_no'] = pd.to_datetime(df_US_pivot['job_posted_month'], format='%B' ).dt.month
df_US_pivot = df_US_pivot.sort_values('job_posted_month_no')
df_US_pivot = df_US_pivot.set_index('job_posted_month')
df_US_pivot = df_US_pivot.drop(columns='job_posted_month_no')

df_US_pivot
```

```python
#now lets plot this.

df_US_pivot.plot(kind='line')
```

```python
# lets just focus on top 3 roles 

top_3 = df['job_title_short'].value_counts().head(3)

top_3
```

```python
top_3 = top_3.index.to_list()
```

```python
df_US_pivot[top_3].plot(kind='line')

plt.title('Monthly Job Postings for Top Data Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```

```python
# #above example in easier method
# an easier way to sort the months. using the .to_period('M') method.

df_US['job_posted_month'] = df_US['job_posted_date'].dt.to_period('M')

df_US
```

```python
df_US_pivot = df_US.pivot_table(index='job_posted_month', columns='job_title_short', aggfunc='size')

df_US_pivot
```

```python
df_US_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend()
plt.show()
```
